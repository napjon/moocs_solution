### To checked for linearity

```r
plot(ml$residuals ~ mlb11$at_bats)
abline(h=0, lty=3)
```

## Nearly normal residuals

```r
hist(ml$residuals)
```

```r
qqnorm(ml$residuals)
qqline(ml$residuals)
```
